```python
max_val = max(df['checkpoint_avg_similarity'])
min_val = min(df['checkpoint_avg_similarity'])
range_val = max_val - min_val

# Calculate percentage difference
# Using the standard formula: (max - min) / min * 100
percentage_difference = (range_val / min_val) * 100

print(f"Max: {max_val:.6f}")
print(f"Min: {min_val:.6f}")
print(f"Range: {range_val:.6f}")
print(f"Percentage difference: {percentage_difference:.2f}%")
```

```python
max_val = max(df['vanilla_avg_similarity'])
min_val = min(df['vanilla_avg_similarity'])
range_val = max_val - min_val

# Calculate percentage difference
# Using the standard formula: (max - min) / min * 100
percentage_difference = (range_val / min_val) * 100

print(f"Max: {max_val:.6f}")
print(f"Min: {min_val:.6f}")
print(f"Range: {range_val:.6f}")
print(f"Percentage difference: {percentage_difference:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))

# Set up positions for bars
x_pos = np.arange(len(df))
width = 0.35

# Create bars
checkpoint_vals = df['checkpoint_avg_similarity']
vanilla_vals = df['vanilla_avg_similarity'] 

checkpoint_bars = plt.bar(x_pos - width/2, checkpoint_vals, 
                         width, label='New Model', alpha=0.7, color='blue')
vanilla_bars = plt.bar(x_pos + width/2, vanilla_vals, 
                      width, label='Old Model', alpha=0.7, color='orange')

# Add mean lines
plt.axhline(y=checkpoint_vals.mean(), color='blue', linestyle='--', linewidth=1.5, label='New Mean')
plt.axhline(y=vanilla_vals.mean(), color='orange', linestyle='--', linewidth=1.5, label='Old Mean')

# Customize the chart
plt.xlabel('States', fontsize=12)
plt.ylabel('Average Similarity', fontsize=12)
plt.title('Average Similarity to "an Indian person" - Checkpoint vs Vanilla Model', fontsize=14)
plt.xticks(x_pos, df['state'], rotation=45, ha='right')
plt.legend()
plt.grid(True, alpha=0.3)

# Add value labels on top of bars
for i, (c_val, v_val) in enumerate(zip(checkpoint_vals, vanilla_vals)):
    plt.text(i - width/2, c_val + 0.002, f'{c_val:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i + width/2, v_val + 0.002, f'{v_val:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Compute relative change
df['rel_change_pct'] = (-df['vanilla_avg_similarity'] +df['checkpoint_avg_similarity']) 

# Sort by relative change
df_sorted = df.sort_values(by='rel_change_pct', ascending=False)

# Print output
for _, row in df_sorted.iterrows():
    print(row['state'], row['checkpoint_avg_similarity'], row['vanilla_avg_similarity'], row['rel_change_pct'])
```
